```python

class NLPCloudClient:

    def __init__(self,api_key):
        """
        Initializes the API client with an API key and sets up the session.
        """
        # Assign the API key to an instance attribute for future use
        self.api_key = api_key
        self.connect=nlpcloud.Client("paraphrase-multilingual-mpnet-base",api_key)
    def paraphrase_text(self, text):
        # 3. Use the toolbelt to make the API call
        try:
            response = self.conne.paraphrasing(text)
            return response
        except Exception as e:
            print(f"API Error: {e}")
            return None
   
```

```python
class ParaphraseResult:
    def __init__(self, raw_dictionary):
        # Dig into the messy dictionary and extract only what we care about
        if raw_dictionary and "paraphrased_text" in raw_dictionary:
            self.text = raw_dictionary["paraphrased_text"]
        else:
            self.text = "Error: Could not extract paraphrased text."
```

```python
class TextManager:
    def __init__(self, api_service):
        # Store the worker so the Manager can use it
        self.api_service = api_service

    def rewrite_paragraph(self, original_text):
        print("Sending text to NLP Cloud...")
        
        # 1. Ask the Worker to get the raw data
        raw_data = self.api_service.paraphrase_text(original_text)
        
        # 2. Package that raw data into our clean Data Model
        clean_result = ParaphraseResult(raw_data)
        
        # 3. Return the final, clean object
        return clean_result
```

```python
# 1. Create the worker (using the key from Cell 1)
my_api_worker = NLPCloudClient(api)

# 2. Create the manager, handing it the worker
my_manager = TextManager(my_api_worker)

# 3. Define the data
target_sentence = "John Doe started learning Javascript when he was 15 years old. After a couple of years he switched to Python."

# 4. Execute the logic!
final_output = my_manager.rewrite_paragraph(target_sentence)

# 5. Display the result
print("\n--- Results ---")
print(f"Original: {target_sentence}")
print(f"Paraphrased: {final_output.text}")
```
