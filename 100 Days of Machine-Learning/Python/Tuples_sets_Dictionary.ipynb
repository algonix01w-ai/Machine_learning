# Difference between Lists and Tuples

<ol>
    <li>syntax</li>
    <li>Mutability</li>
    <li>Speed - Tuples are faster than list</li>
    <li>Memory</li>
    <li>Built in Functionlity</li>
    <li>Error Prone</li>
    <li>Usability</li>
</ol>

```python
import time 
L = list(range(10000000))
T = tuple(range(10000000))

start = time.time()
for i in L:
    i*4
print('List time', time.time()-start)
start = time.time()
for i in T:
    i*4
print('Tuple time', time.time()-start)
```

```python
import sys
L = list(range(1000))
T = tuple(range(1000))

print('List size',sys.getsizeof(L))
print('Tuple size', sys.getsizeof(T))
```

# Sets

A set is an unordered collection of items. Every set element is unique (no duplicates) and must be immutable (cannot be changed).

However, a set itself is mutable. We can add or remove items from it.

Sets can also be used to perform mathematical set operations like union, intersection, symmetric difference, etc.

Characterstics:
- Unordered
- Mutable
- No Duplicates
- Can't contain mutable data types

```python
# empty 
s={}
print(s)
print(type(s))
S=set()
print(S)
print(type(S))
#1D and 2D
s1={1,2,3}
print(s1)
# s2={1,2,3,4,{5,6}}#TypeError: unhashable type: 'set'
# print(s2)

#Homo and Hetro
s3= {1,'hello',2.6,True}
print(s3)# true not print bcoz in set not duplicates (True == 1)
s4= {1,'hello',2.6,(5,6,8,1)}
print(s4)
#using type conversion
s5=set([1,2,5,5,9,2,8,5])
print(s5)

#duplicates not allowed
# set can't have mutable items
s6={1,2,[3,4]}
print(s6)
```

```python
s={}
print(s)
print(type(s))
S=set(1)
print(S)
print(type(S))
```

```python
s1={1,2,3}
s2={2,1,3}
print(s1==s2)
```

```python
# len/sum/min/max/sorted
s={2,3,5,6,8,9}
print(len(s))
print(min(s))
print(max(s))
print(sorted(s))
sum(s)
```

# Dictionary

Dictionary in Python is a collection of keys values, used to store data values like a map, which, unlike other data types which hold only a single value as an element.

In some languages it is known as map or assosiative arrays.

dict = { 'name' : 'Sanjana' , 'age' : 3, 'gender' : 'Female' }

Characterstics:

- Mutable
- Indexing has no meaning
- keys can't be duplicated
- keys can't be mutable items

```python
s = {
    'name':'Saloni',
     'college':'bit',
     'sem':3,
     'subjects':{
         'dsa':50,
         'maths':67,
         'english':34
     }
}
s
```

```python
for  key in s:
    print(key,s[key])
```

```python
for  value in s.values():
    print(value)
```
