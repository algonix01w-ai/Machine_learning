# File Handling


### Some Theory

##### Types of data used for I/O:
- Text - '12345' as a sequence of unicode chars
- Binary - 12345 as a sequence of bytes of its binary equivalent

##### Hence there are 2 file types to deal with
- Text files - All program files are text files
- Binary Files - Images,music,video,exe files

### How File I/O is done in most programming languages

- Open a file
- Read/Write data
- Close the file

```python
# readline () -> to read line by line
f=open('/Machine_learning/100 Days of Machine-Learning/Python/sample1.txt','r')
print(f.readline(),end="")
print(f.readline())
print(f.readline())
f.close()
```

```python
# reading entire using readline
f= open('/Machine_learning/100 Days of Machine-Learning/Python/sample1.txt','r')
while True:
    data = f.readline()

    if data=='':
        break
    else:
        print(data,end="")
f.close()
```

### Using Context Manager (With)

- It's a good idea to close a file after usage as it will free up the resources
- If we dont close it, garbage collector would close it
- with keyword closes the file as soon as the usage is over

```python
#with
with open('/Machine_learning/100 Days of Machine-Learning/Python/sample1.txt','w') as f:
    f.write('muzaffarpur institute of technology\n')
    f.write('technology')
```

```python
# try f.read() now
with open('/Machine_learning/100 Days of Machine-Learning/Python/sample1.txt','r') as f:
    print(f.read())
```

```python
# moving within a file  -> 10 char then 10 char
with open('sample1.txt','r') as f:
    print(f.read(10))
    print(f.read(10))

```

```python
# benefits -> to load a i big file in memory
big_L = ['hello world ' for i in range(1000)]

with open('big.txt','w') as f:
    f.writelines(big_L)
```

```python
with open('big.txt','r') as f:
    chunk_size = 100

    while len(f.read(chunk_size))>0:
        print(f.read(chunk_size),end="****")
        f.read(chunk_size)
```

```python
#seek and tell function
with open('sample1.txt','r') as f:
    print(f.read(10))
    print(f.tell())
```

```python

with open('sample1.txt','r') as f:
    print(f.read(10))
    print(f.tell())
    f.seek(0)
    print(f.read(10))
    print(f.tell())
    f.seek(25)
    print(f.read(10))
```

```python
# seek during write
with open('sample1.txt','w') as f:
    f.write('HELLO')
    f.seek(1)
    f.write('XASJ')
```
