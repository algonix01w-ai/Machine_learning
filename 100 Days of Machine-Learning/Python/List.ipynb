# <i><b>List Comprehension</b></i>

<p>List comprehension provides a concise way of creating lists</p>
<h3>newlist =[expression <b><i>for</i></b> item <b><i>in</i></b> iterable if condition == True]<h3>

## Advantages of List Comprehension

<ul>
    <li>More time-effcient and space-effcient than loops.</li>
    <li>Require Fewer lines of code.</li>
    <li>Transforms iterative statment into a formula</li>
</ul>

```python
# Add 1 to 10 numbers to a list 
l=[]
for i in range(1,11):
    l.append(i)
print(l)
    
```

```python
a=[i for i in range(1,12)]
print(a)
```

```python
# scalar multipication on a vector
v=[2,3,6]
s=-3
x=[]
for i in v:
    x.append(i*s)
print(x)
```

```python
# in one line
y=[s*i for i in v ]
print(y)
```

```python
#add squares
L=[1,2,3,4,5,6]
z=[i*i for i in L]
print(z)
```

```python
[i for i in range(1,51) if i%5==0]
```

```python
languages=['Python','java','go','r','c','c++','php','pph']
[lan for lan in languages if lan.startswith('p') ]
```

```python
[lan for lan in languages if lan.startswith('p') if lan.endswith("p") ]
```

```python
[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
l1=[1,2,3]
l2=[3,2,1]
[i*j for i in l1 for j in l2]
```

# Zip

<p>The zip() function returns a zip object which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together.</p>
<p>If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator</p>

```python
# write a program to add items of 2 lists indexwise
L1=[1,2,3]
L2=[3,-5,2,1,-8,-6]
# list(zip(L1,L2))

[i+j for i,j in zip(L1,L2) ]
```
