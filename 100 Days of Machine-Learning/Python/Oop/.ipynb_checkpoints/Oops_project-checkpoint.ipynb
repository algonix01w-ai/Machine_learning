```python

class NLPCloudClient:

    def __init__(self,api_key):
        """
        Initializes the API client with an API key and sets up the session.
        """
        # Assign the API key to an instance attribute for future use
        self.api_key = api_key
        self.connect=nlpcloud.Client("paraphrase-multilingual-mpnet-base-v2",api_key)
    def text_p(self,text):
        response=self.connect.paraphrasing(text)
        return response
        
   
```
